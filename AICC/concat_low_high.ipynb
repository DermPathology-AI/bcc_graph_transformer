```python
from pathlib import Path
import pandas as pd
import glob
```

```python
save_path = '../datasets/bcc_concat/data/'
Path(save_path).mkdir(parents=True, exist_ok=True)
file_paths = glob.glob('../datasets/bcc_low/data/*')
file_names = [file.split('/')[-1] for file in file_paths]
```

```python
for file_name in file_names:
    low = pd.read_csv(f'../datasets/bcc_low/data/{file_name}')
    high = pd.read_csv(f'../datasets/bcc_high/data/{file_name}')
    low_high = pd.concat([low, high], axis=1)
    low_high.to_csv(f'../datasets/bcc_concat/data/{file_name}', header=None, index=False)
```
