```python
import pandas as pd
import numpy as np
from scipy import stats
```

```python
def ensemble_acc(classes):
    m0  = pd.read_csv(f'../val_test_results/GT_double_{classes}C_0F_test.csv')
    m1  = pd.read_csv(f'../val_test_results/GT_double_{classes}C_1F_test.csv')
    m2  = pd.read_csv(f'../val_test_results/GT_double_{classes}C_2F_test.csv')
    m3  = pd.read_csv(f'../val_test_results/GT_double_{classes}C_3F_test.csv')
    m4  = pd.read_csv(f'../val_test_results/GT_double_{classes}C_3F_test.csv')
    m = pd.DataFrame()

    m['preds_0'] = m0['preds']
    m['preds_1'] = m1['preds']
    m['preds_2'] = m2['preds']
    m['preds_3'] = m3['preds']
    m['preds_4'] = m3['preds']
    
    m['majority'] = stats.mode(np.array(m.T))[0].T
    
    Acc_m0 = round(len(m0[m0['label']==m0['preds']])/len(m0), 3)
    Acc_m1 = round(len(m1[m1['label']==m1['preds']])/len(m1), 3)
    Acc_m2 = round(len(m2[m2['label']==m2['preds']])/len(m2), 3)
    Acc_m3 = round(len(m3[m3['label']==m3['preds']])/len(m3), 3)
    
    print('Acc m0:', round(len(m0[m0['label']==m0['preds']])/len(m0), 3))
    print('Acc m1:', round(len(m1[m1['label']==m1['preds']])/len(m1), 3))
    print('Acc m2:', round(len(m2[m2['label']==m2['preds']])/len(m2), 3))
    print('Acc m3:', round(len(m3[m3['label']==m3['preds']])/len(m3), 3))
    
    print()
    
    acces = np.array([Acc_m0, Acc_m1, Acc_m2, Acc_m3])*100
    
    print('Mean Acc:', round (np.mean(acces),3), np.std(acces))
    print('Acc ensemble:',round(len(m0[m0['label']==m['majority']])/len(m), 3))
    print()
    return m
```

```python
ensemble_acc(5) # change 2, 3, or 5
```
