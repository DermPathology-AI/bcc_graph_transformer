```python
import os
import matplotlib.pyplot as plt
import glob
import numpy as np
path = '/workspace/data/cv_methods/tmi2022/WSI/BCC43 (3)_1a_files/20.0/'
all_image = glob.glob(path + '/*')
w = 10
h = 10
fig = plt.figure(figsize=(10, 10))
columns = 20
rows    = 20
for i in range(1, columns*rows +1):
    img = plt.imread(all_image[i])
    if np.mean(img) >100:
        fig.add_subplot(rows, columns, i)
        plt.imshow(img)
    plt.axis('off')
plt.show()
```
