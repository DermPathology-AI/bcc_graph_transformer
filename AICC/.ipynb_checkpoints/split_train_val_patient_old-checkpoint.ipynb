```python
def get_id(bcc_codes):
    patient_id = list(set([bcc for bcc_code in bcc_codes for bcc in re.findall('^BCC\d+', bcc_code) if len(bcc)!=0]))
    return patient_id[0]
```

```python
def ndpi_2_files(ndpi):
    return ndpi[0].replace('.ndpi', '_files/20.0')
```

```python
wsi_label_patient_id['patient id']   = wsi_label_patient_id['bcc code'].apply(lambda x: get_id([x]))
wsi_label_patient_id['tiles folder'] = wsi_label_patient_id['names'].apply(lambda x: ndpi_2_files([x]))
wsi_label_patient_id
```

```python
def split_patients(bcc_annotation):
    bcc_codes = list(bcc_annotation['bcc code'])
    patient_id = list(set([bcc for bcc_code in bcc_codes for bcc in re.findall('^BCC\d+', bcc_code) if len(bcc)!=0]))
    X_train, X_test = model_selection.train_test_split(pd.DataFrame(patient_id), test_size=0.1, random_state=2023)
    return X_train, X_test

def get_patient_ids(bcc_annotation):
    bcc_codes = list(bcc_annotation['bcc code'])
    return list(set([bcc for bcc_code in bcc_codes for bcc in re.findall('^BCC\d+', bcc_code) if len(bcc)!=0]))
    
def get_file_names_patient(bcc_code):
    bcc_files = []
    for file in glob.glob('/workspace/data/bcc-projekt-digital/bcc/*/*'):
        bcc_files.append(file.split('/')[-1])
        
    bcc_set=[]
    for file in tqdm(bcc_files):
        for xt in list(bcc_code['names']):
            if len(re.findall(f'^{xt}', file)): 
                bcc_set.append(file)
    return bcc_set


def annotation_graph_trans_binary_patient(X_code, file_save_name, grades=2):
    bcc_annotation = file_names_2_df_patient(path, get_file_names_patient(X_code))
    display(bcc_annotation)
    all_bcc = glob.glob('/workspace/data/cv_methods/tmi2022/WSI/*')
    folders = []
    for bcc in all_bcc:
        if bcc.endswith('.dzi'):pass
        else: folders.append(bcc)

    wsi_labels = []
    for folder in folders:
        f = folder.replace('_files','').split('/')[-1]

        try:
            label = float(bcc_annotation[bcc_annotation['names'] == f + '.ndpi']['label'])
            wsi_labels.append((folder.split('/')[-1] + '/20.0', int(label)))
        except Exception as e:
            #print(e, folder)
            pass
    
    wsi_label_df = pd.DataFrame(wsi_labels, columns=['X', 'y'] )
     
    X = wsi_label_df['X']
    y = wsi_label_df['y'].astype(int)

    if grades==1:
        y_ = [1 if yt!=0 else 0 for yt in list(y)]
    elif grades==2:
        y_  = []
        for yt in list(y):
            if yt==0:
                y_.append(0)
            elif yt==1 or yt==2:
                y_.append(1)
            elif yt==3 or yt==4:
                y_.append(2)
    else:
        y_ = y
        
    data_set = pd.DataFrame()
    #val_set = pd.DataFrame()
    data_set['X'] = X
    data_set['y'] = y_
    #val_set['X']   = X_test
    #val_set['y']   = y_test_

    graph_trans_script_path = '/workspace/data/cv_methods/tmi2022/scripts/'

    data_set.to_csv(graph_trans_script_path + f'{file_save_name}', sep='\t', index=False, header=False)

    #val_set.to_csv(graph_trans_script_path + 'val_set.txt', sep='\t',     index=False, header=False)
    return data_set
```
