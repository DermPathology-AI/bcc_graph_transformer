# Metrics

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import ConfusionMatrixDisplay, confusion_matrix
from sklearn.metrics import roc_curve, roc_auc_score, precision_recall_fscore_support, classification_report
from sklearn.model_selection import KFold
from sklearn.metrics import roc_curve, auc
from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression
plt.style.use('seaborn')
```

## Generate ROC curves of multiple classes

```python
def get_senitivity_specifity(val_set_df):
    cm=confusion_matrix(val_set_df['label'], val_set_df['preds'])
    FP = cm.sum(axis=0) - np.diag(cm)  
    FN = cm.sum(axis=1) - np.diag(cm)
    TP = np.diag(cm)
    TN = cm.sum() - (FP + FN + TP)
    TPR = TP/(TP+FN)
    TNR = TN/(TN+FP) 
    print('Senitivity:', TPR)
    print('Specifity:',  TNR)
    return TPR, TNR

def multi_label_roc(labels, predictions,proba, num_classes, pos_label=1, fold=0):
    fprs = {}
    tprs = {}

    thresholds = []
    thresholds_optimal = []
    aucs = []
    precisions = []
    recalls = []
    fscores = []
    roc_auc = dict()
    base_fpr = np.linspace(0, 1, 101)
   
        
    if len(predictions.shape)==1:
        predictions = predictions[:, None]
        
    for c in range(0, num_classes):        
        label = labels[:, c]
        prediction = predictions[:, c]
        p = proba[:,c]
        fpr, tpr, threshold = roc_curve(label, p)
        
        c_auc = roc_auc_score(label, p)
        aucs.append(c_auc)
        thresholds.append(threshold)
 

        if num_classes==2 and c==0:
            continue
        plt.plot(list(fpr),list(tpr))
        
        plt.ylabel('True Positive Rate')
        plt.xlabel('False Positive Rate')

         
        tpr_m = np.interp(base_fpr, fpr, tpr)
        tpr_m[0] = 0.0
 

        fprs[str(c) + '_fpr_'  + str(fold) + '_F' ]= list(base_fpr)
        tprs[str(c) + '_tpr_'  + str(fold) + '_F' ]= list(tpr_m)
    
    
    precision, recall, fscore, _ = precision_recall_fscore_support(label, prediction, average='binary')
    precisions.append(precision)
    recalls.append(recall)
    fscores.append(fscore)
    
    TPR_SS, TNR_SS = get_senitivity_specifity(val_set_df)
    
    if num_classes == 5:
        plt.legend([f'Class 0, AUC: {round(aucs[0],3)}', 
                    f'Class 1, AUC: {round(aucs[1],3)}',
                    f'Class 2, AUC: {round(aucs[2],3)}',
                    f'Class 3, AUC: {round(aucs[3],3)}',
                    f'Class 4, AUC: {round(aucs[4],3)}'])
    if num_classes == 3:
        plt.legend([f'Class 0, AUC: {round(aucs[0],3)}', 
                    f'Class 1, AUC: {round(aucs[1],3)}',
                    f'Class 2, AUC: {round(aucs[2],3)}'])
    if num_classes == 2:
        plt.legend([f'Class 1, AUC : {round(aucs[1],3)}'
                    ])

    print(classification_report(labels, predictions))
    return aucs, precisions, recalls, fscores, TPR_SS, TNR_SS, fprs, tprs
```

```python
TNRs = []
TPRs = []
f = 0
classes = 3
fprs_models = {}
tprs_models = {}

k=0
for f in range(0, 4):
    f = 0
    plt.figure()
    val_set_name = f'GT_double_{classes}C_{f}F'
    val_set_df  = pd.read_csv(f'../val_set_results/{val_set_name}.csv')
    labels      = np.array(pd.get_dummies(val_set_df['label']))
    predictions = np.array(pd.get_dummies(val_set_df['preds']))
    proba = val_set_df['proba']

    proba_float=[]

    for row in  proba:
        proba_float.append(list(map(float,np.array(row.strip(']').strip('[').split(',')))))
    proba = np.array(proba_float) 
    
    k+=1

    aucs, precisions, recalls, fscores, TPR, TNR,  fprs, tprs = multi_label_roc(labels=labels, predictions=predictions, proba=proba, num_classes=classes, fold=k)
        
    fprs_models.update(fprs)
    tprs_models.update(tprs)
    
    TPRs.append(TPR)
    TNRs.append(TNR)

    plt.plot([0, 1], [0, 1], linestyle='--', lw=2, label='Random', color='b', alpha=.8)
    plt.savefig(f'figures/roc_{val_set_name}.png', bbox_inches='tight',pad_inches=0.1)
    plt.savefig(f'figures/roc_GT_proba_{classes}C_{f}F.png', bbox_inches='tight',pad_inches=0.1)
    print('TNRs:', np.round(np.mean(np.array(TNRs), axis=0), 3), np.round(np.std(np.array(TNRs), axis=0), 3))
    print('TPRs:', np.round(np.mean(np.array(TPRs), axis=0), 3), np.round(np.std(np.array(TPRs), axis=0), 3))   
```

## Generate mean of ROC curves of mulitple models

```python
def draw_mean_roc(tprs_models, num_classes = 3):
    
    def mean_class(mean_tprs_m, line_color = 'b'):
        plt.plot(base_fpr, mean_tprs_m, line_color)
        
    def std_class(mean_tprs_m,std,line_color = 'b'):
        tprs_upper = np.minimum(mean_tprs_m + std, 1)
        tprs_lower = mean_tprs_m - std
        plt.fill_between(base_fpr, tprs_lower, tprs_upper, color=line_color, alpha=0.3)
    
    line_colors = ['b','g', 'r', 'm', 'y']
    base_fpr = np.linspace(0, 1, 101)
    tprs_models = pd.DataFrame(tprs_models)
    tprs_model_mean = {}
    tprs_model_std = {}
    aucs = []
    
    for i in range(num_classes):
        c = str(i)
        tprs_model_mean[str(i)] = list(tprs_models.loc[:,[col  for col in tprs_models.columns if col.startswith(c)]].mean(axis=1))
        tprs_model_std[str(i)]  = list(tprs_models.loc[:,[col  for col in tprs_models.columns if col.startswith(c)]].std(axis=1))
        mean_class(np.array(tprs_model_mean[c]), line_color  = line_colors[i])
        aucs.append(auc(base_fpr, np.array(tprs_model_mean[c])))

    plt.plot([0, 1], [0, 1],'b--')    
    plt.xlim([-0.01, 1.01])
    plt.ylim([-0.01, 1.01])
    plt.ylabel('True Positive Rate')
    plt.xlabel('False Positive Rate')
        
    if num_classes == 5:
        plt.legend([f'Class 0, AUC: {round(aucs[0],3)}', 
                    f'Class 1, AUC: {round(aucs[1],3)}',
                    f'Class 2, AUC: {round(aucs[2],3)}',
                    f'Class 3, AUC: {round(aucs[3],3)}',
                    f'Class 4, AUC: {round(aucs[4],3)}'])
    if num_classes == 3:
        plt.legend([f'Class 0, AUC: {round(aucs[0],3)}', 
                    f'Class 1, AUC: {round(aucs[1],3)}',
                    f'Class 2, AUC: {round(aucs[2],3)}'])
    if num_classes == 2:
        plt.legend([f'Class 1, AUC : {round(aucs[1],3)}'
                    ])
     
    for i in range(num_classes):
        c = str(i)
        tprs_model_mean[str(i)] = list(tprs_models.loc[:,[col  for col in tprs_models.columns if col.startswith(c)]].mean(axis=1))
        tprs_model_std[str(i)]  = list(tprs_models.loc[:,[col  for col in tprs_models.columns if col.startswith(c)]].std(axis=1))
        std_class(np.array(tprs_model_mean[c]), np.array(tprs_model_std[c]) + 0.1, line_color  = line_colors[i])
    plt.savefig(f'figures/roc_mean_std_{classes}_C.png', bbox_inches='tight',pad_inches=0.1)
draw_mean_roc(tprs_models,  num_classes = 3)
```

## Training progress

```python
def read_plot_accuracy(double=False,clas=2, is_all_in_one=False):
    plt.figure(figsize=(15,10))
    all_acc = []
    for fold in range(5):
        if double: file_path = f'/workspace/data/cv_methods/tmi2022/graph_transformer/runs/GT_double_{clas}C_{fold}F_{clas}.log' 
        else: file_path = f'/workspace/data/cv_methods/tmi2022/graph_transformer/runs/GT_{clas}C_{fold}F_{clas}.log' 
            
        with open(file_path) as f:
            f = f.readlines()
        tvs=[['0', '0', '0']]

        for tv in f:
            tvs.append(tv.replace('\n','').split(' '))

        plt.subplot(3,2,fold+1)
        
        df_tv_score = pd.DataFrame(np.array(tvs)).astype(float)
        df_tv_score.iloc[:,1].plot()
        df_tv_score.iloc[:,2].plot()
        acc = round(np.max(df_tv_score.iloc[:,2]),3)
        all_acc.append(acc)
        plt.xlabel('Epoch')
        plt.ylabel('Accuracy')
        plt.legend(['Training','Validation'])
        plt.title(f'Fold # {fold}, Acc: {acc}')
        plt.grid('off')
    print('Overall accuracy', np.max(all_acc))
    print('Mean acc', np.mean(all_acc), np.std(all_acc))
    return all_acc
```

### Single Magnification

```python
classes = 3
all_accs = read_plot_accuracy(double=False, clas=classes)
```

### Double magnification

```python
all_double_accs = read_plot_accuracy(double=True, clas=classes)
```

## Single vs Double Magnification

```python
dict_acc = dict(zip(['0','1','2','3','4'], all_accs))
dict_double_acc = dict(zip(['0','1','2','3','4'],all_double_accs))
two_acc = pd.DataFrame()
two_acc['Single magnificaiton'] = pd.DataFrame(dict_acc, index=[0]).T
two_acc['double magnification'] = pd.DataFrame(dict_double_acc, index=[0]).T
```

```python
plt.figure()
ax = two_acc.plot.bar(figsize=(10,10))
for container in ax.containers:
    ax.bar_label(container)
plt.xlabel('Fold #')
plt.ylabel('Accuracy')
plt.title(f'Validation set, {classes}-class')
plt.savefig(f'figures/single_double_compare_{classes}.png', bbox_inches='tight',pad_inches=0.1)
```

### Training progress of models

```python
all_acc_tasks = pd.DataFrame()
for classes in [2, 3, 5]:
    try:
        all_accs= read_plot_accuracy(double=False, clas=classes)
        all_double_accs = read_plot_accuracy(double=True, clas=classes)
    except Exception as e:
        print(e)
        
    dict_acc = dict(zip(['0','1','2','3','4'], all_accs))
    dict_double_acc = dict(zip(['0','1','2','3','4'],all_double_accs))
    all_acc_tasks[f'2 magn, {classes} classes'] = pd.DataFrame(dict_double_acc, index=[0]).T
    all_acc_tasks[f'1 magn, {classes} classes'] = pd.DataFrame(dict_acc, index=[0]).T
 
```
