```python
def extend_graph(org_g, new_g):
    sz_org_g = np.shape(org_g)[0]
    sz_new_g = np.shape(new_g)[0]
    size_combined =  sz_org_g + sz_new_g
    
    combined_g = np.zeros((size_combined,size_combined))
    combined_g[:sz_org_g, :sz_org_g] = org_g[:sz_org_g, :sz_org_g]
    combined_g[sz_org_g:, sz_org_g:] = new_g[:sz_new_g, :sz_new_g]
    return combined_g
```
